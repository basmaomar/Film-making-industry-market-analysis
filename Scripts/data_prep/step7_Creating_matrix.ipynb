# Create two cumulative matrices, one for person/ nominated and the other for person/ won .

```python
import pandas as pd
import numpy as np
import ast
```

```python
df = pd.read_csv('awardsdict3.csv')
df.shape
```

```python
indx = ['name', 2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019]
len(indx)
```

```python
#df['index'] = indx
df.head()
```

```python
df.info()
```

```python
win = df.loc[1:]
nom = df.loc[1:]
```

```python
for i in win['0230859']:
    print(ast.literal_eval(i))
```

### Create won matrix

```python
win_col =  win.columns


for w in win_col:
    win[w] = win[w].apply(lambda col : col.count('Winner'))
win.head()
```

### Create nominated matrix

```python
nom_col =  nom.columns
for n in nom_col:
    nom[n] = nom[n].apply(lambda col : col.count('Nominee'))
nom.head()
```

```python
winner = win.transpose()
```

```python
winner['id']= winner.index
wn = list( df.loc[0])
winner['i']= wn
winner.set_index('i')
```

```python
winner.head()
```

```python
nominate = nom.transpose()
```

```python
nominate.head()
```

```python
nominate['id']= winner.index
nm = list( df.loc[0])
nominate['i']= nm
nominate.set_index('i')
```

### Save matrices as CSV files

```python
export_csv = winner.to_csv (r'winners.csv', index = None, header=True)
```

```python
export_csv = nominate.to_csv (r'nominee.csv', index = None, header=True)
```
