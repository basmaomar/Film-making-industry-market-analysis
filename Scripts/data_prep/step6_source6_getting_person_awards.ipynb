# Getting the awards of each person in each year by webscraping www.imdb.com

```python
import requests
import random
import re
import pandas as pd
from bs4 import BeautifulSoup
```

```python
id_name = pd.read_csv('d3.csv') 
```

```python
id_name.columns
```

```python
ids_list = id_name['ID'] # replace with id lists

awardsdict={} 

for i in ids_list: # for each actor
    cid = i
    link = 'https://www.imdb.com/name/nm'+str(i) +'/awards?ref_=nm_ql_2'
    print(link)
    response = requests.get(link)
    print(response.status_code)
    thisaward = [] 

    soup = BeautifulSoup(response.text, 'html.parser')  
    tt1 = soup.find_all('table', class_ = 'awards')
    for tt in tt1:
        y = tt.find('a')
        aa = tt.find('td', class_="award_outcome").find('b')
        #print(y, a)
        thisaward.append((int(y.text.replace('\n','').replace(' ', '')),aa.text))
    awardsdict[cid] = thisaward
    sortedAwards = {2000:[],2001:[],2002:[],2003:[],2004:[],2005:[],2006:[],2007:[],2008:[],2009:[],2010:[],2011:[],2012:[],2013:[],2014:[],2015:[],2016:[],2017:[],2018:[],2019:[]}

    for a in awardsdict[cid]:
        for y in range (2000,2019):
            if a[0] < y:
                sortedAwards[y].append(a[1])
    awardsdict[cid] = sortedAwards
    
```

```python
#save into CSV file
pd.DataFrame(awardsdict).to_csv('awardsdict.csv',index=False,header=True)
```
