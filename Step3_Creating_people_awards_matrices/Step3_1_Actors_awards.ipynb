### UnderStanding the data

```python
Movies_.head(2)
```

```python
# take top 2 people
tcast = []

for c in Movies_['cast']:
    tempc=[]
    try:
        tempc = c.split(',')
        tcast.append(tempc[0].replace('[','').replace('\'',''))
        tcast.append(tempc[1].replace('[','').replace('\'',''))
    except:
        continue
```

```python
len(tcast)
```

```python
#remove duplicatiion
cast = set(tcast)
len(cast)
```

```python
#create a dataframe
cast_info = pd.DataFrame()
#add people names
cast_info['Name'] = list(cast)
```

```python
# get people ID's and add to the dataframe
IDs = []
for n in cast_info['Name']:
    try:
        cid = moviesDB.search_person(n)[0].getID()
        print(cid)
        IDs.append(cid)
    except:
        IDs.append(None)
cast_info['ID'] = IDs
```

cast_info.head(5)

```python
#save into CSV file
cast_info.to_csv('cast_info_new_2009_200.csv',index=False,header=True)
```
